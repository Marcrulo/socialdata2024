> * Create a barplot of crimes-per-year (years on the $x$-axis, crime-counts on the $y$-axis).
> * Finally, Police chief Suneman is interested in the temporal development of only a **subset of categories, the so-called *focus crimes***. Those categories are listed below (for convenient copy-paste action). Create bar-charts displaying the year-by-year development of each of these categories across the years 2003-2017.
> 

> <i>Answers</i>

```python
import numpy as np
grouped = df['Category'].groupby(df['Year']).value_counts().unstack().replace(np.nan, 0)
```

```python
group_sum = grouped.sum(axis=1)
print(2003+group_sum.argmax())
print(group_sum.max())
print()
print(2003+group_sum.argmin())
print(group_sum.min())
```

```python
plt.bar(group_sum.index, group_sum.values)
plt.xlabel("Year")
plt.ylabel("Frequency (count)")
plt.title("Frequency of crime over the years")
plt.xticks(grouped.index,rotation=90)
plt.show()
```

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
crime_list = sorted(list(focuscrimes))
```

```python
grouped
```

```python
n_row = 7
n_col = 2
fig, axs = plt.subplots(n_row,n_col,figsize=(15,20))
for row in range(n_row):
    for col in range(n_col):
        axs[row, col].bar(grouped.index, grouped[crime_list[row+7*col]])
        axs[row, col].set_title(crime_list[row+7*col])
        if row == n_row-1:
            axs[row, col].set_xticks(grouped.index)
            for tick in axs[row, col].get_xticklabels():
                tick.set_rotation(90)
        else:
            axs[row, col].set_xticks([])


fig.tight_layout()
plt.show()
```
